Do you know that there is now a well-known game League of Legends. It holds some tournaments every year, and in China the name of this tournament is called LPL, which is similar to NBA. I counted the play of some famous top players in lpl. Their ids are, summit, theshy, rich and flandre. I used these data to make a mod. this mod can help some clubs to get the players they want. For example, when a rookie top player finishes a match, input his data into this mod and you can get his style of play. If it shows "theshy", then it means that the player is very aggressive on the field and can be the core of the team. If it shows "rich", then it means that the player can still produce high output in team battles even if he is not given too much resources. And if a club needs a top player who can help the other menbers team during the game, then they can choose the player with result "flandre". And if they want a top player with a magnificent kda, then "summit" is a quality option.

download the lpl top file from this link https://drive.google.com/file/d/1n1_sr0VCKbg-UQIUoqubyneMR0ITnZAm/view?usp=sharing



```python
from sklearn import svm

import pandas as pd

df = pd.read_csv("lpl top.csv")

df
```

set kill, death, assistance, damage dealt, damage taken, gold earned as x  

```python
X = df[['K', 'D', 'A', 'Damage dealt', 'Damege taken', 'Gold earned']]
X
```

set player's style as y

```python
Y = df[['Style']]
Y
```

Encode player's style


```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

yEncoded = le.fit_transform(Y['Style'])
```

train model

```python
irisPredictionModel = svm.SVC()

irisPredictionModel.fit(X, yEncoded)
```

input: place to enter player's performence 

```python
prediction = irisPredictionModel.predict([[ 7, 0, 5, 20000, 23000, 20143]] )

```

output: player's style

```python
returnToOrginal = le.inverse_transform(prediction)

print(returnToOrginal)
```
